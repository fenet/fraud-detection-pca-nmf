Eigenfaces: Facial Feature Extraction and Representation (Using a Subset of Olivetti Dataset)

Subject IDs: 1, 10, 29, 33, 38

Objective: To Implement PCA from scratch (Eigenfaces) on this subset, compare with Scikit-learn's PCA, explain the curse of dimensionality, and apply Non-Negative Matrix Factorization (NMF) to this face data subset. 

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import MinMaxScaler
```

```python
ds = np.load('olivetti_subset.npz')
X_all = ds['images'] 
y_all = ds['labels']
fixed_subject_ids = (1, 10, 29, 33, 38)

mapp = {1: 0, 10: 1, 29: 2, 33: 3, 38: 4}
y = np.array([mapp[val] for val in y_all])

plt.figure(figsize=(10, 2))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    tid = fixed_subject_ids[i]
    loc = np.where(y_all == tid)[0][0]
    plt.imshow(X_all[loc].reshape(64, 64), cmap='gray')
    plt.title("ID: " + str(tid))
plt.show()
```

```python
X_all = X_all.reshape(len(X_all), -1)
# 1. Calculate the mean face
mean_face = np.mean(X_all, axis=0)

# 2. Center the data
X_centered = X_all - mean_face

# 3. Perform SVD
U, s, Vt = np.linalg.svd(X_centered, full_matrices=False)
eigenfaces_scratch = Vt

# 4. Visualize the top eigenfaces
plt.figure(figsize=(10, 2))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(eigenfaces_scratch[i].reshape(64, 64), cmap='bone')
    plt.title("Face " + str(i+1))
    plt.axis('off')
plt.show()
# 5. Project the centered data onto the first two principal components
X_pca_scratch = np.dot(X_centered, Vt[:2].T)

X_centered = X_all - np.mean(X_all, axis=0)
```

```python
pca = PCA(n_components=2, random_state=69)
X_pca_sklearn = pca.fit_transform(X_all)
# Project to 2D for plotting using sklearn PCA
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_pca_scratch[:, 0], X_pca_scratch[:, 1], c=y)
plt.title("Minilab PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.subplot(1, 2, 2)
plt.scatter(X_pca_sklearn[:, 0], X_pca_sklearn[:, 1], c=y)
plt.title("sklearn PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
   
```

```python
# Apply NMF to the 5-subject face subset (X,y)
X_flat = X_all.reshape(len(X_all), -1)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_flat)
nmf = NMF(n_components=2, init='random', random_state=69, max_iter=1000)
X_nmf = nmf.fit_transform(X_scaled)
# TODO: Create scatter plot for NMF
plt.figure(figsize=(7, 5))
plt.scatter(X_nmf[:, 0], X_nmf[:, 1], c=y)
plt.title("NMF Projection")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.colorbar(label='Subject ID (Mapped)')
plt.show()
# ...
```
